```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('/Users/apple/Downloads/new_data_actual.csv')
df.head()
```

```python
df.hist(bins=15, color='red', edgecolor='black', linewidth=1.0,
           xlabelsize=8, ylabelsize=8, grid=False)    
plt.tight_layout(rect=(0, 0, 2, 2))
```

```python
# Histogram
fig = plt.figure(figsize = (6,4))
title = fig.suptitle("Histogram for Temp (Deg)", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax = fig.add_subplot(1,1, 1)
ax.set_xlabel("temperature")
ax.set_ylabel("Frequency") 
freq, bins, patches = ax.hist(df['Temp (Deg)'], color='steelblue', bins=15,
                                    edgecolor='black', linewidth=1)
                                    

# Density Plot
fig = plt.figure(figsize = (6, 4))
title = fig.suptitle("bivariate kernel density estimate.", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=10)

ax1 = fig.add_subplot(1,1, 1)
ax1.set_xlabel("temperature")
ax1.set_ylabel("Frequency") 
sns.kdeplot(df['Temp (Deg)'], ax=ax1, shade=True, color='steelblue')
```

```python
ax = sns.kdeplot(df['Temp (Deg)'],df.index ,cmap="Reds", shade=True, shade_lowest=False)
```

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10, 6))
corr = df.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=0.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Wine Attributes Correlation Heatmap', fontsize=20)
```

```python
Depth1=df['Depth']
minimum_Depth = np.amin(Depth1)
maximum_Depth = np.amax(Depth1)
mean_Depth = np.mean(Depth1)
median_Depth = np.median(Depth1)
std_Depth = np.std(Depth1)
print("Statistics for Depth Coloumn\n")
print("Minimum Depth: {}".format(minimum_Depth)) 
print("Maximum Depth: {}".format(maximum_Depth))
print("Mean Depth: {}".format(mean_Depth))
print("Median Depth {}".format(median_Depth))
print("Standard deviation of Depth: {}".format(std_Depth))
```

```python
WS1=df['WS']
minimum_WS1 = np.amin(WS1)
maximum_WS1 = np.amax(WS1)
mean_WS1 = np.mean(WS1)
median_WS1 = np.median(WS1)
std_WS1 = np.std(WS1)
print("Statistics for ws Coloumn\n")
print("Minimum WS: {}".format(minimum_WS1)) 
print("Maximum WS: {}".format(maximum_WS1))
print("Mean WS: {}".format(mean_WS1))
print("Median WS {}".format(median_WS1))
print("Standard deviation of WS: {}".format(std_WS1))
```
